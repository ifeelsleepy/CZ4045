```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
def get_data(filename):
    with open(filename, 'r') as f:
        data = f.read()
    data_list =data.split(',')
    ppl_float=[]
    for i in range(len(data_list)-1):
        ppl_str = data_list[i][-9:].strip()
        ppl_float.append(float(ppl_str))
    return ppl_float
```

```python
normal_ppl = get_data('train_data_normal2.txt')
sgd_ppl = get_data('train_data_Adam1.txt')
tied_ppl = get_data('train_data_tied2.txt')
tied_ppl_sgd = get_data('train_data_tied_Adam1.txt')
```

```python
plt.plot(normal_ppl, label='Normal Perplexity Score')
plt.plot(sgd_ppl, label='Perplexity Score Using Adam')
plt.plot(tied_ppl, label='Perplexity Score with tied weights')
plt.plot(tied_ppl_sgd, label='Perplexity Score with tied weights and Adam')
plt.legend()
```
